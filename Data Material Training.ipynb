Taking a look on our html we see that we have a lot of tables have single constant values. (Meaning they are booleans - Exists or not).

```python
# First let's take a 
df_material['ADDITIVES_BRANDED_DIESEL_IND'].unique() 
# Two possibilities Y - Yes, nan -Not a number
```

```python
# Creating filters for diesel industries where t
diesel_ind = df_material['ADDITIVES_BRANDED_DIESEL_IND'] == 'Y'
not_diesel_ind = df_material['ADDITIVES_BRANDED_DIESEL_IND'] != 'Y'

# Using the filters replace the values 
df_material.loc[diesel_ind,['ADDITIVES_BRANDED_DIESEL_IND']] = True
df_material.loc[not_diesel_ind,['ADDITIVES_BRANDED_DIESEL_IND']] = False #or df[~diesel_ind] = 1
```

```python
# Checking our change
df_material['ADDITIVES_BRANDED_DIESEL_IND'].unique()
```

```python
# Creating a function to that
# Replacing values in a "text" boolean column
# Does the same think as we did before
def replace_val_column(df,old_value,new_value,columns,fill_value=None): 
    """Function that replace a value on the data frame on an range of columns 
    
    Arguments:
        df  -- Pandas DataFrame
        old_value  -- value to be replaced on the dataframe
        new_value  -- value that will replace 
        columns  -- columns that you want to use
        fill_value  -- value that will replace de NaN
    """
    # Function applymap that maps each cell and to the logic
    df[columns] = df[columns].applymap(lambda val: new_value if old_value == val else val)
    if fill_value is not None:
        df[columns] = df[columns].fillna(fill_value)
    return df
```

```python
# Gotten from the analysis
columns_bools_name = ["ADDITIVES_BRANDED_GASOLINE_IND","ADDITIVES_DYE_IND","ADDITIVES_LUBRICITY_IND","ADDITIVES_UNBRANDED_GASOLINE_IND"]
df_material = replace_val_column(df=df_material,old_value='Y',new_value=True,columns=columns_bools_name,fill_value=False)
```

### Creating a new column 

Using the transformations that we did we will create a new column

```python
# Replacing the bool values for the name of the column
def replace_bool_for_header(df,columns):
    for column_name in columns:
        # Loc function that receives a filter and the column name
        df.loc[df[column_name] == True,column_name] = column_name
    return df
```

```python
columns_to_join = [header for header in df_material.columns.values if header.startswith('ADDITIVES')]
df_material = replace_bool_for_header(df=df_material,columns=columns_to_join)
columns_to_join
```

```python
# Creating a new column "Industry" using the apply function to use the the row joinner function created
def row_joinner(row):
    # Filtering values which has no
    return ','.join(filter(lambda val: isinstance(val,str),row.values))

df_material['Additive'] = df_material[columns_to_join].apply(lambda row: row_joinner(row) , axis=1)
```

### Treating the data

One of the biggest problems that I have had here is the bad quality data texts that don't have patterns, double spaces, dashes where it should be spaces, anyway a lot of problems here we gonna do a small treatment of data.

```python
def treat_string_data(df):
    # Removing spaces and double from headers 
    df.columns = df.columns.str.strip().str.replace('  ',' ')
    # Removing spaces and replacing duplicate spaces for single spaces.
    # Text columns are considere of type "object" that's why the if val.dtype == object
    return df.apply(lambda val: val.str.strip().str.replace('\s+',' ',regex=True).str.upper() if val.dtype == object else val)

# Treating the Text data (strings) 
df_material = treat_string_data(df_material)
```

```python
# Separate the columns that don't add any value to the analysis
def separate_important_values(df):
    # Getting the column values
    headers = df.columns.values
    const_columns = {header:df[header].unique() for header in headers if len(df[header].unique()) <= 1}
    good_columns = [header for header in headers if len(df[header].unique()) > 1]
    return const_columns,good_columns

# Finding columns that add and don't add values
const_columns,good_col = separate_important_values(df_material)
```
